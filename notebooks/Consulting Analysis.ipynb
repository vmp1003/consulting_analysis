```python
import pandas as pd 
```

```python
df_contracts = pd.read_csv('../data/contracts.csv')
df_contracts.head()
```

```python
df_customers = pd.read_csv('../data/customers.csv')
df_customers.head()
```

```python
df_customers.rename(columns={'id': 'customer_id'}, inplace= True)
df_customers.head()
```

```python
df_customers.set_index('customer_id', inplace = True)
df_customers.head()
```

```python
df_contracts = df_contracts.set_index('customer_id')
df_contracts.head()
```

```python
df = df_contracts.join(df_customers, how='left')
df.head(25)
```

```python
df = df_contracts.join(df_customers, how='left')
df.head(25)

df['deposit'] = pd.to_numeric(df['deposit'], errors='coerce')
```

```python
df = df_contracts.join(df_customers, how='left')
df.head(25)

df['monthyl_amt'] = pd.to_numeric(df['monthly_amt'], errors='coerce')

```
